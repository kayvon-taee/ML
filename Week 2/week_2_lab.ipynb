```python
# PLot output of matplotlib as graphic not as a separate window (Use %matplotlib alone for separate window)
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

irisnf = pd.read_csv("irisnf.csv")
plt.style.use("seaborn-v0_8")
irisnf
```

```python
plt.plot("Petal.l","Sepal.l",data=irisnf)
```

```python
plt.plot("Petal.l","Sepal.l",data=irisnf, linestyle="none",marker="o")
```

```python
plt.plot("Petal.l","Sepal.l",data=irisnf, linestyle="none",marker="o", alpha=0.5, markersize=10)
```

```python
plt.plot("Petal.l","Sepal.l",data=irisnf, linestyle="none",marker="o", alpha=0.5, markersize=10, markeredgecolor="black",markerfacecolor="red")
plt.xlabel("Petal Length")
plt.ylabel("Sepal Length")
plt.xlim(0,7)
plt.ylim(3,10)
plt.xticks([])
plt.yticks([])
plt.title("Scatterplot")
```

```python
# c = colour by key
# s = adjust size by Sepal.w
# Lab sheet uses irisnf["Sepal.w"].values*20
# Using .values uses more memory because it converts it to a numpy array
# MatPlotLib can use both numpy array and pandas series, so fine not to use .values
plt.scatter("Petal.l","Sepal.l",data=irisnf,c="Variety",s=irisnf["Sepal.w"]*20, alpha=0.6)
```

```python
plt.scatter("Petal.l","Sepal.l",data=irisnf, c="Variety")
plt.savefig("plot1.png")
```

```python
p=sns.scatterplot(x="Petal.l",y="Sepal.l",marker="",data=irisnf) 
for line in range(0,irisnf.shape[0]): 
    l=irisnf.iloc[line] 
    p.text(l["Petal.l"],l["Sepal.l"],int(l["Variety"]))
```

```python
p=sns.scatterplot(x="Petal.l",y="Sepal.l",data=irisnf)
p.figure.suptitle("Plot of lengths", fontsize=16)
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Sepal length", fontsize=14)
sns.pairplot(irisnf)
sns.pairplot(irisnf).savefig("plot2.png")
```

```python
# Initialise the facet grid, col=column
p=sns.FacetGrid(data=irisnf, col="Variety")
# Choose the plot type, x and y axes
p.map(plt.scatter,"Petal.l","Sepal.l").add_legend()
```

```python
# Simultaneously plot two variables together
sns.jointplot(x="Petal.l",y="Sepal.l",data=irisnf)
# Same but contour plot
sns.jointplot(x="Petal.l",y="Sepal.l",data=irisnf,kind="kde")
"""
Contour plots use the KDE method, which is an abbreviation for 'kernel density estimation.' 
This is where we put a Gaussian distribution at each point and take a contour map of the result.
"""
```
